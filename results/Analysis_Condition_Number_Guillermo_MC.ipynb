```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data_mu1 = pd.read_csv("condition_numbers_mu_0.1.csv").set_index("Trial")
data_mu2 = pd.read_csv("condition_numbers_mu_0.01.csv").set_index("Trial")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_elegant_histograms_with_means(data_mu1, data_mu2, mu1=0.1, mu2=0.01):
    mean_mu1 = data_mu1.mean().values[0]
    mean_mu2 = data_mu2.mean().values[0]
    
    plt.style.use('seaborn')
    
    plt.figure(figsize=(10, 6))
    plt.hist(data_mu1, bins=30, alpha=0.5, color='skyblue', edgecolor='darkblue', linewidth=1.5, label=f'μ = {mu1}')
    plt.axvline(mean_mu1, color='red', linestyle='--', linewidth=2, label=f'Mean (μ = {mu1}): {mean_mu1:.2f}')
    plt.xlabel('Condition Numbers', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title(f'Histogram of Condition Numbers for μ = {mu1} (Guillermo Dataset)', fontsize=14, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()
    

    plt.figure(figsize=(10, 6))
    plt.hist(data_mu2, bins=30, alpha=0.5, color='lightgreen', edgecolor='darkgreen', linewidth=1.5, label=f'μ = {mu2}')
    plt.axvline(mean_mu2, color='red', linestyle='--', linewidth=2, label=f'Mean (μ = {mu2}): {mean_mu2:.2f}')
    plt.xlabel('Condition Numbers', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title(f'Histogram of Condition Numbers for μ = {mu2} (Guillermo Dataset)', fontsize=14, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()
```

```python
plot_elegant_histograms_with_means(data_mu1, data_mu2)
```
