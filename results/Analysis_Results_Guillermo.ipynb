```python
import pandas as pd
import matplotlib.pyplot as plt
```

# This notebook plots the replication results of the Guillermo Dataset

```python
data = pd.read_excel("guillermo_replication_results.xlsx")
```

```python
def plot_deff_rank(df):

    mu_vals = df['mu_vals']
    d_eff = df['d_eff']
    rank_mean = df['rank_mean']
    
    plt.figure(figsize=(8, 4))
    plt.plot(mu_vals, d_eff, label=r'$d_{\mathrm{eff}}(\mu)$')
    plt.plot(mu_vals, rank_mean, label=r'$\ell$  mean')
    plt.axhline(y=4200, color='r', linestyle='--', label=r'$\ell_{max}$')

    plt.xscale('log')

    plt.gca().invert_xaxis()

    plt.xlabel('μ')
    plt.ylabel('Rank')
    #plt.title('Effective Dimension and Rank Mean over Regularization Parameter μ')
    plt.title(r'Effective dimension $d_{\mathrm{eff}}(\mu)$ and approximate mean rank over $\mu$')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_deff_rank(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_cg_pcg_times(df):

    mu_vals = df['mu_vals']
    time_cg_mean = df['time_cg_mean']
    time_pcg_mean = df['time_pcg_mean']
    time_cg_std = df['time_cg_std']
    time_pcg_std = df['time_pcg_std']

    plt.figure(figsize=(8, 4))

    plt.plot(mu_vals, time_cg_mean, label="CG Time", color="blue")
    plt.fill_between(mu_vals, time_cg_mean - time_cg_std, time_cg_mean + time_cg_std, color="blue", alpha=0.2)

    plt.plot(mu_vals, time_pcg_mean, label="Nys-PCG Time", color="red")
    plt.fill_between(mu_vals, time_pcg_mean - time_pcg_std, time_pcg_mean + time_pcg_std, color="red", alpha=0.2)

    plt.xscale("log")
    #plt.yscale("log")
    plt.gca().invert_xaxis()

    plt.xlabel("Regularization Parameter (μ)")
    plt.ylabel("Time (s)")
    plt.title("CG vs. Nys-PCG Computation Times over μ ")
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)

    plt.show()
```

```python
plot_cg_pcg_times(data)
```
